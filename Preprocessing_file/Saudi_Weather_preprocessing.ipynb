##Preprocessing 

### Extract Day, Month & hour Separately from observation_date column

```python
#Define the format of observation_date column
df['observation_date'] = pd.to_datetime(df.observation_date, format='%Y-%m-%d %H:%M:%S')
```

```python
# Remove the timezone information from the observation_date column
df['observation_date']=df['observation_date'].dt.tz_localize(None)
```

```python
df['observation_date'].value_counts()
```

```python
#Extracting 'Month' separately from observation_date column
df['month'] = df['observation_date'].dt.month
```

```python
df['month'].value_counts()
```

```python
#Extracting 'Day' separately from observation_date column
df['day'] = df['observation_date'].dt.day
```

```python
df['day'].value_counts()
```

```python
#Extracting 'hour' separately from observation_date column
df['hour'] = df['observation_date'].dt.hour
```

```python
df['hour'].value_counts()
```

```python
#Extracting 'minute' separately from observation_date column
df['minute'] = df['observation_date'].dt.minute 
```

```python
df['minute'].value_counts()
```

### Create 'Season' column from Month

```python
#Create 'Season' column from month
df['season'] = (df.month - 1) // 3
```

Mapping month 1,2,3 to 0 (winter season),
month 4,5,6 to 1 (spring season),
month 7,8,9 to 2 (summer season), and month
10,11,12 to 3 (fall season).

```python
df['season'].value_counts()
```

```python
#Create 'Season_name' column from month
season_dict = {1: 'Winter',
               2: 'Winter',
               3: 'Winter', 
               4: 'Spring',
               5: 'Spring',
               6: 'Spring',
               7: 'Summer',
               8: 'Summer',
               9: 'Summer',
               10: 'Fall',
               11: 'Fall',
               12: 'Fall'}
df['Season_name'] = df['month'].apply(lambda x: season_dict[x])
```

```python
df['Season_name'].value_counts()
```

## Create 'humidity' column from dew point

```python
#Create 'humidity_level' column from air_temperature_dew_point column
def my_func(dew):
  if dew <= 10:
        return '25% and lower'
  elif dew <= 12:
        return '26â€“30%'
  elif dew <= 16:
        return'31â€“36%'
  elif dew <= 18:
        val = '37â€“43%'
  elif dew <= 21:
        return '44â€“51%'
  elif dew <= 24:
        return '52â€“61%'
  elif dew <= 26:
        return '62â€“72%'
  else:
        val = '73% and higher'
  return val
df['humidity_level'] = df['air_temperature_dew_point'].apply(my_func)
```

```python
df['humidity_level'].value_counts()
```

### Create air_temperature_categories from air_temperature

```python
#Create 'air_temperature_categories' column from 'air_temperature' column
def my_func(temp):
  if temp<= 15:
        return 'cold'
  elif (temp > 15) & (temp<= 23) :
        return'comfortable'
  elif (temp > 23) & (temp <= 32):
        val = 'warm'
  else:
        val = 'hot'
  return val
df['air_temperature_categories'] = df['air_temperature'].apply(my_func)
```
