### >> Send to consumer station name with comfortable atmosphere in summer season . And just show the station name , air temperature, Season name and air temperature categories columns 
****

```python
from kafka import KafkaAdminClient, KafkaConsumer, KafkaProducer
from kafka.admin import NewTopic
import json
from json import loads
import csv
from csv import DictReader

# Required setting for Kafka Producer
topicname = 'comfortable_atmosphere'
producer = KafkaProducer(bootstrap_servers='localhost:9092',value_serializer=lambda v: binascii.hexlify(v.encode('utf-8')))
producer = KafkaProducer()


# iterate over each line as a ordered dictionary and print only few column by column name
with open('SaudiWeather.csv','r') as read_obj:
    csv_dict_reader = DictReader(read_obj,dialect=csv.excel)
    for column in csv_dict_reader:
        if column["air_temperature_categories"] == 'comfortable' and column["Season_name"] == 'Summer':
            ack = producer.send(topicname,json.dumps((column['station_name'] ,column['air_temperature'],column["Season_name"],column['air_temperature_categories'])).encode('utf-8'))
            metadata = ack.get()
            print(metadata.topic, metadata.partition)
            
            
```
